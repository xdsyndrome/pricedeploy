```python
# Import libraries
import joblib
import os
import pandas as pd
from sklearn.model_selection import train_test_split
import lightgbm as lgb
```

```python
df = pd.read_csv("./data/df_treatment.csv")
```

```python
# Clean dataset
df_reg = df[['floor_area_sqm', 'town', 'remaining_lease', 'resale_price', 'storey_category', 'distance_to_mrt', 'distance_to_pri_sch', 'distance_to_shopping', 'treatment', 'post', 'intervention']]
df_reg = pd.get_dummies(df_reg)

for col in df_reg.columns:
    if '/' in col or '-' in col or ' ' in col:
        new_col_name = col.replace('/', '')
        new_col_name = new_col_name.replace('-', '')
        new_col_name = new_col_name.replace(' ', '')
        df_reg.rename(columns={col: new_col_name}, inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_reg.drop(columns=['resale_price']), 
                                                    df_reg['resale_price'], 
                                                    test_size=0.25,
                                                    random_state=1
                                                   )
```

```python
hyper_params = {
    'task': 'train',
    'boosting_type': 'gbdt',
    'objective': 'regression',
    'metric': ['l1','l2'],
    'learning_rate': 0.005,
    'feature_fraction': 0.9,
    'bagging_fraction': 0.7,
    'bagging_freq': 10,
    'verbose': 0,
    "max_depth": 8,
    "num_leaves": 128,  
    "max_bin": 512,
    "num_iterations": 100000
}
```

```python
lgbm = lgb.LGBMRegressor(**hyper_params)
lgbm.fit(X_train, y_train, eval_set=[(X_test, y_test)], eval_metric='l1', early_stopping_rounds=500)
```

```python
# Save the ML model
if not os.path.exists("model"):
    os.makedirs("model")

joblib.dump(lgbm, "model/lgbm.pkl")
```
